```python
%%writefile app.py
import streamlit as st
import pandas as pd
from io import StringIO
from pathlib import Path
import torch
import torchvision.models as models

# Init model
from pathlib import Path
# import requests
from io import BytesIO
from PIL import Image
# from functools import cache

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import torchvision.transforms as T
import torch


if "model" not in st.session_state.keys():
    model = torch.load('runs/train/exp/weights/best.pt')
    st.session_state["model"] = model


st.set_page_config(
        page_title="XRay Diagnosis",
)

st.title("Chẩn đoán XRay lồng ngực")
input_Xray = st.file_uploader("Chọn ảnh khuôn mặt bạn")



if st.button("Diagonosis!!!", type="primary"):
    # st.image('result.png', caption='Sunrise by the mountains')

    if input_Xray is not None:
        # To read file as bytes:
        save_folder = '/content/gdrive/MyDrive/upload'
        face_path = Path(save_folder, input_Xray.name)
        with open(face_path, mode='wb') as w:
            w.write(input_Xray.getvalue())
        # Save output to file ressult.png
        save_path = Path(save_folder, 'result.png')
        final_image = model(face_path)
        final_image = T.functional.to_pil_image(final_image)
        # final_image.save(save_path)
        st.image(final_image, caption='result')
        del final_image
```
